```python
import pandas as pd
```

```python
df=pd.DataFrame({'team':['a','a','a','b','b','b','b','c','c','c'],'points':[25,26,25,11,23,43,24,64,76,33]})
```

```python
df
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['team']=le.fit_transform(df['team'])
```

```python
df
```

# onehot encoding


```python
df=pd.DataFrame({'team':['a','a','a','b','b','b','b','c','c','c'],'points':[25,26,25,11,23,43,24,64,76,33]})
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown='ignore')
df_new=pd.DataFrame(ohe.fit_transform(df[['team']]).toarray())
df_new
```

```python
#merge df with new_df 
df=df.join(df_new)
df
```

```python
df=df.drop('team',axis=1)
df
```

```python
df.columns=['points','teamA','teamB','teamC']
df
```
